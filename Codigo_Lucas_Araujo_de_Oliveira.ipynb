```python

# 1.3 Qual o pedido mais extenso, qual o seu tamanho?

import pandas as pd

# Carregar o arquivo .xlsx 
df = pd.read_excel('dados.xlsx')

# Inicializar variáveis para armazenar o pedido mais extenso e seu tamanho
pedido_mais_extenso = ""
tamanho_pedido_mais_extenso = 0

# Iterar pelos pedidos para encontrar o mais extenso e calcular seu tamanho
for index, row in df.iterrows():
    pedido = row['Pedido']
    tamanho_pedido = 0
    
    for caracter in pedido:
        tamanho_pedido += 1
    
    if tamanho_pedido > tamanho_pedido_mais_extenso:
        pedido_mais_extenso = pedido
        tamanho_pedido_mais_extenso = tamanho_pedido

# Exibir o pedido mais extenso e seu tamanho
print("O pedido mais extenso é:")
print(pedido_mais_extenso)
print(f"  Tamanho: {tamanho_pedido_mais_extenso} caracteres")
```

```python
#2.3 Qual é o tempo médio de resposta? 
import pandas as pd

# Carregar o arquivo .xlsx
df = pd.read_excel('dados.xlsx')

# Inicializar variáveis
soma_tempo_resposta = 0
contagem_tempo_resposta = 0

# Iterar pelas linhas do DataFrame e somar os tempos de resposta
for index, row in df.iterrows():
    tempo_resposta = row['Dias para Resposta']
    soma_tempo_resposta += tempo_resposta
    contagem_tempo_resposta += 1

# Calcular o tempo médio de resposta
tempo_medio_resposta = soma_tempo_resposta / contagem_tempo_resposta

# Exibir o resultado
print("Tempo médio de resposta:", tempo_medio_resposta)
```

```python
#2.4 Qual a moda do tempo médio de resposta? 
import pandas as pd

# Carregar o arquivo .xlsx 
df = pd.read_excel('dados.xlsx')

# lista para armazenar os tempos de resposta
tempos_resposta = []

# Iterar pelas linhas do DataFrame e adicionar os tempos de resposta à lista
for index, row in df.iterrows():
    tempo_resposta = row['Dias para Resposta']
    tempos_resposta.append(tempo_resposta)

# Calcular a moda
frequencia_tempo_resposta = {}
moda_tempo_resposta = []

for tempo in tempos_resposta:
    frequencia = tempos_resposta.count(tempo)
    if frequencia > 1 and tempo not in moda_tempo_resposta:
        moda_tempo_resposta.append(tempo)

# Exibir o resultado
print("Moda do tempo médio de resposta:", moda_tempo_resposta)
```

```python
#2.5 Qual a mediana do tempo médio de resposta? 
import pandas as pd

# Carregar o arquivo .xlsx 
df = pd.read_excel('dados.xlsx')

# lista para armazenar os tempos de resposta
tempos_resposta = []

# Iterar pelas linhas do DataFrame e adicionar os tempos de resposta à lista
for index, row in df.iterrows():
    tempo_resposta = row['Dias para Resposta']
    tempos_resposta.append(tempo_resposta)

# Ordenar a lista de tempos de resposta
tempos_resposta.sort()

# Calcular a mediana
tamanho_lista = len(tempos_resposta)
if tamanho_lista % 2 == 0:
    mediana = (tempos_resposta[tamanho_lista // 2 - 1] + tempos_resposta[tamanho_lista // 2]) / 2
else:
    mediana = tempos_resposta[tamanho_lista // 2]

# Exibir o resultado
print("Mediana do tempo médio de resposta:", mediana)

```

```python
#3.2 Porcentagem de cada identidade de gênero em relação ao total de pedidos
import pandas as pd

# Carregar o arquivo .xlsx 
df = pd.read_excel('dados.xlsx')

# Contar a frequência de cada gênero
contagem_identidades_genero = df['Identidade de gênero'].value_counts()

# Calcular o total de pedidos
total_pedidos = len(df)

# Calcular e exibir a porcentagem de cada gênero
print("Porcentagem de cada identidade de gênero em relação ao total de pedidos:")
for identidade_genero, contagem in contagem_identidades_genero.items():
    porcentagem = (contagem / total_pedidos) * 100
    print(f"   . {identidade_genero}: {porcentagem:.2f}%")

```

```python
#4.1 Liste os 5 Estados com maior número de pedidos de acesso à informação, juntamente com a quantidade correspondente, em ordem decrescente 

import pandas as pd

# Carregar o arquivo .xlsx
df = pd.read_excel('dados.xlsx')

# Criar um dicionários para contar os pedidos por Estado
contagem_estados = {}

# Iterar sobre as linhas do DataFrame
for index, row in df.iterrows():
    estado = row['Estado']
    if estado in contagem_estados:
        contagem_estados[estado] += 1
    else:
        contagem_estados[estado] = 1

# Ordenar o dicionários  com base no número de pedidos em ordem decrescente
estados_ordenados = sorted(contagem_estados.items(), key=lambda x: x[1], reverse=True)

# Pegar os 5 Estados com o maior número de pedidos
cinco_estados = estados_ordenados[:5]

# Exibir os 5 Estados 
print("Os 5 Estados com maior número de pedidos de acesso à informação:")
for estado, quantidade in cinco_estados:
    print(f"  . {estado}: {quantidade} pedidos")
```

```python
#4.2 Quais cidadãos possuem o maior número de pedidos de acesso à informação, liste os 3 mais frequentes e a quantidade de pedidos de cada 
import pandas as pd

# Carregar o arquivo .xlsx 
df = pd.read_excel('dados.xlsx')

#  add contagem de pedidos por cidadão
contagem_cidadaos = {}

# Iterar pelas linhas do DataFrame para contar os pedidos por cidadão
for index in range(len(df)):
    cidadao = df.loc[index, 'Nome do Cidadão']
    if cidadao in contagem_cidadaos:
        contagem_cidadaos[cidadao] += 1
    else:
        contagem_cidadaos[cidadao] = 1

# Ordenar a lista em ordem decrescente de quantidade de pedidos e pegar os 3 primeiros
top_3_cidadaos = sorted(contagem_cidadaos, key=contagem_cidadaos.get, reverse=True)[:3]

# Exibir os 3 cidadãos com o maior número de pedidos
print("Os 3 cidadãos com o maior número de pedidos de acesso à informação:")
for cidadao in top_3_cidadaos:
    quantidade_pedidos = contagem_cidadaos[cidadao]
    print(f"   . {cidadao}: {quantidade_pedidos} pedidos")
```

```python

# 5.1 Há alguma correlação entre o tempo de resposta e alguma outra coluna?
import pandas as pd

# Carregar o arquivo .xlsx 
df = pd.read_excel('dados.xlsx')

# Inicializar variáveis para calcular a correlação 
n = len(df)  # Número de observações
soma_x = 0  # (tempo de resposta)
soma_y = 0  #(tamanho do pedido)
soma_xy = 0  # Soma dos produtos de x e y
soma_x_quadrados = 0  # Soma dos quadrados de x
soma_y_quadrados = 0  # Soma dos quadrados de y

# Criar uma nova coluna com o tamanho de cada pedido
df['Tamanho do Pedido'] = df['Pedido'].apply(lambda x: len(x))

# Iterar pelas linhas do DataFrame para calcular as somas
for index, row in df.iterrows():
    tempo_resposta = row['Dias para Resposta']
    tamanho_pedido = row['Tamanho do Pedido']
    
    soma_x += tempo_resposta
    soma_y += tamanho_pedido
    soma_xy += tempo_resposta * tamanho_pedido
    soma_x_quadrados += tempo_resposta ** 2
    soma_y_quadrados += tamanho_pedido ** 2

# Calcular a correlação 
correlacao = (n * soma_xy - soma_x * soma_y) / ((n * soma_x_quadrados - soma_x ** 2) * (n * soma_y_quadrados - soma_y ** 2)) ** 0.5

# Exibir o resultado
print(f"Correlação de Pessoas: {correlacao:.2f}")
```

```python
# 5.2 Caso exista, quais seriam as entradas para um modelo preditivo para o tempo de resposta?

import pandas as pd

# Carregar o arquivo .xlsx
df = pd.read_excel('dados.xlsx')

# Criar uma nova coluna com o tamanho de cada pedido
df['Tamanho do Pedido'] = df['Pedido'].apply(len)

# Definir recursos )
X = df['Tamanho do Pedido']
y = df['Dias para Resposta']

# Calcular os coeficientes da regressão linear
n = len(X)
soma_X = 0
soma_y = 0
soma_Xy = 0
soma_X_quadrado = 0

for i in range(n):
    soma_X += X[i]
    soma_y += y[i]
    soma_Xy += X[i] * y[i]
    soma_X_quadrado += X[i] ** 2

b1 = (n * soma_Xy - soma_X * soma_y) / (n * soma_X_quadrado - soma_X ** 2)
b0 = (soma_y - b1 * soma_X) / n

# Fazer previsões para o conjunto de dados
previsoes = [b0 + b1 * x for x in X]

# Exibir os resultados
for i in range(n):
    print(f"Pedido {i+1}: Tamanho do Pedido = {X[i]}, Tempo de Resposta Previsto = {previsoes[i]:.2f}")
```

```python
#6.Questão Extra (Desafio): Conte as palavras nos pedidos de acesso à informação, por exemplo: 
import pandas as pd

# Carregar o arquivo .xlsx 
df = pd.read_excel('dados.xlsx')

# Criar uma nova coluna para armazenar o número de palavras
df['Palavras no Pedido'] = 0

# Iterar pelas linhas do DataFrame e contar as palavras nos pedidos
for index, row in df.iterrows():
    pedido = row['Pedido']
    palavras = []  #  lista vazia para as palavras
    palavra_atual = ''  # string vazia para a palavra atual
    for caracter in pedido:
        # Verificar se o caracter é uma letra ou dígito
        if caracter.isalnum():
            palavra_atual += caracter
        elif palavra_atual:
            # Se a palavra atual não for vazia, add à lista de palavras
            palavras.append(palavra_atual)
            palavra_atual = ''  # 0 a palavra atual
    # Verificar se a última palavra foi encontrada
    if palavra_atual:
        palavras.append(palavra_atual)
    # Atualizar a coluna com o número de palavras
    df.at[index, 'Palavras no Pedido'] = len(palavras)

# mostrar o resultado
print("Contagem de Palavras nos Pedidos:")
print(df[['Pedido', 'Palavras no Pedido']])
```
